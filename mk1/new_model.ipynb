```python
# a function to update the class labels of the new 'aiming-soldier' dataset:

# Path to the 'labels' folder
folder_path_of_labels = "C:\\Users\\kruth\\Datasets\\defence\\soldier_plus_weapon\\train\\labels"

def change_labels(folder_path_of_labels):
    # iterate over each file in the 'labels' directory
    for filename in os.listdir(folder_path_of_labels): 
        file_path = os.path.join(folder_path_of_labels, filename) # get the full path of the file
        
        if filename.endswith('.txt'): # check if the file is a text file
            with open(file_path, 'r') as file: # open the file in read mode
                lines = file.readlines() 

            modified_lines = [] 
            for line in lines: # iterate over each line in the file
                modified_line = line.replace('0', '4') 
                modified_lines.append(modified_line)

            with open(file_path, 'w') as file: # open the file in write mode
                file.writelines(modified_lines) # write the modified lines back to the file

    print("All labels have been updated.")

# implement the function
change_labels(folder_path_of_labels)
```
